```python
def get_topic_page(topic_url):
    response = requests.get(topic_page_url)
    if response.status_code != 200:
        raise expection('error {}'.format(topic_url))
        topic_doc = BeautifulSoup(response.text, 'html.parser')
        return topic_doc


def get_repo_info(h3_tag, star_tag):
    a_tags = h3_tag.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url = base_url + a_tags[1]['href']
    stars = star_tags[0].text.strip()
    return username, repo_name, repo_url, stars


def get_topic_repos(topic_doc):
    response = requests.get(topic_page_url)
    if response.status_code != 200:
        raise expection('error {}'.format(topic_url))
    topic_doc = BeautifulSoup(response.text, 'html.parser')


h3_class = 'f3 color-fg-muted text-normal lh-condensed'
repo_tags = topic_doc.find_all('h3', class_=h3_class)


stars = 'Counter js-social-count'
star_tags = topic_doc.find_all('span', class_=stars)


topic_repos_dict = {
    'username': [],
    'repo_name': [],
    'repo_url': [],
    'stars': [],
}


for i in range(len(repo_tags)):
    repo_info = get_repo_info(repo_tags[i], star_tags[i])
    topic_repos_dict['username'].append(repo_info[0])
    topic_repos_dict['repo_name'].append(repo_info[1])
    topic_repos_dict['repo_url'].append(repo_info[2])
    topic_repos_dict['stars'].append(repo_info[3])
    return pd.DataFrame(topic_repos_dict)


def scrape_topic(topic_url , topic_name):
    topic_df = get_topic_repos(get_topic_page(topic_url))
    topic_df.to_csv(topic_name + .'csv' , index= None)
        
```

```python
def get_topic_titles(doc):
    selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'
    topic_tags = doc.find_all('p' ,class_=selection_class)
    topic_title = []

for tag in topic_tags:
    topic_title.append(tag.text)
    return topic_title


def get_topic_descs(doc):
    description_class = 'f5 color-fg-muted mb-0 mt-1'
    descri_tags = doc.find_all('p', class_=description_class)
    topic_des = []

for tag in descri_tags: 
    topic_des.append(tag.text)

    return topic_des


def get_topic_urls(doc):
    link_class ='no-underline flex-grow-0'
topic_link = doc.find_all('a' , class_=link_class)
    
    topic_url = []
base_url ='https://github.com'

for tag in topic_link:
    topic_url.append(base_url + tag['href'])

    
    
    return topic_url

    


  
    


def scrape_topics():
    topics_url ='https://github.com/topics'
     response = requests.get(topics_url)
         if response.status_code != 200:
        raise expection('error {}'.format(topic_url))
        
        topics_dict ={
            'title':get_topic_titles(doc),
            'description':  get_topic_descs(doc),
            'urls': get_topic_urls(doc)
        }
        return pd.DataFrame(topics_dict)
    
    

    

```

```python
def scrape_topics_repos():
    topics_df = scrape_topics()
    for index, row in topics_df.iterrowa():
        print('scrapping "{}"'.format(row['title']))
    scrape_topic(row['url'], row['title'])
```
